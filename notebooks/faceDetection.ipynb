```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nadia = cv2.imread('Nadia_Murad.jpg',0)
denis = cv2.imread('Denis_Mukwege.jpg',0)
solvay = cv2.imread('solvay_conference.jpg',0)
```

```python
plt.imshow(nadia, cmap='gray')
```

```python
faceCascade = cv2.CascadeClassifier('/content/sample_data/haarcascade_frontalface_default.xml')
```

```python
def detectFace(img):
  faceImg = img.copy()
  faceRects = faceCascade.detectMultiScale(faceImg)
  for (x,y,w,h) in faceRects:
    cv2.rectangle(faceImg, (x,y), (x+w,y+h), (255,255,255),10)
  return faceImg
```

```python
result = detectFace(nadia)
plt.imshow(result, cmap='gray')
```

```python
result = detectFace(denis)
plt.imshow(result, cmap='gray')
```

```python
result = detectFace(solvay)
plt.imshow(result, cmap='gray')
```

```python
eyeCascade = cv2.CascadeClassifier('/content/sample_data/haarcascade_eye.xml')
```

```python
def detectEyes(img):
  faceImg = img.copy()
  eyes = eyeCascade.detectMultiScale(faceImg)
  for (x,y,w,h) in eyes:
    cv2.rectangle(faceImg, (x,y), (x+w,y+h), (255,255,255),10)
  return faceImg
```

```python
result = detectEyes(nadia)
plt.imshow(result, cmap='gray')
```

```python
result = detectEyes(denis)
plt.imshow(result, cmap='gray')
```

```python
result = detectEyes(solvay)
plt.imshow(result, cmap='gray')
```

```python
from google.colab.patches import cv2_imshow

# Load the cascades for left and right eyes
leftEyeCascade = cv2.CascadeClassifier('/content/sample_data/haarcascade_lefteye_2splits.xml')
rightEyeCascade = cv2.CascadeClassifier('/content/sample_data/haarcascade_righteye_2splits.xml')

# Function to detect eyes
def detectEyes(img):
  # Check if the image is already grayscale and convert if necessary
  if len(img.shape) == 2: # Grayscale image
    gray = img
  else:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  # Detect left eyes
  leftEyes = leftEyeCascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 5, minSize = (30, 30))
  for (x,y,w,h) in leftEyes:
    cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0),2)

  # Detect right eyes
  rightEyes = rightEyeCascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 5, minSize = (30, 30))
  for (x,y,w,h) in rightEyes:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0),2)

  return img

# Load image from file, ensuring it's in color
img = cv2.imread('Nadia_Murad.jpg', cv2.IMREAD_COLOR)

# Detect eyes in the image
outputImage = detectEyes(img)

# Display the output image
cv2_imshow(outputImage)

# Display output
cv2_imshow(outputImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Load image from file, ensuring it's in color
img = cv2.imread('Denis_Mukwege.jpg', cv2.IMREAD_COLOR)

# Detect eyes in the image
outputImage = detectEyes(img)

# Display the output image
cv2_imshow(outputImage)

# Display output
cv2_imshow(outputImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
